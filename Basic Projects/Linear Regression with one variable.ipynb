# House Price Prediction - Linear Regression with one variable

## Goal

We’ll use linear regression to predict the price of a house given its size.
Our task is to find a relationship between these two variables so that we can estimate prices for new, unseen house sizes.

### What is Linear Regression?

Linear Regression is a method used to model the relationship between a dependent variable (output) and one or more independent variables (inputs).

When we have only one input variable, the algorithm is called Univariate Linear Regression.

“Uni” means one, and “variate” means variable.
So, Univariate Linear Regression = Linear Regression with one feature/variable.

## Terminology

Before we start coding, let’s understand some basic terms used in Machine Learning.
| Term                            | Symbol                 | Meaning                                                                                               |
| ------------------------------- | ---------------------- | ----------------------------------------------------------------------------------------------------- |
| **Training set**                | —                      | The dataset used to train the model. It contains both input features and corresponding output labels. |
| **Input variable (Feature)**    | ( x )                  | The variable used as input to predict the output. Example: size of the house.                         |
| **Output variable (Target)**    | ( y )                  | The actual output or value we want to predict. Example: price of the house.                           |
| **Number of training examples** | ( m )                  | Total number of data points (rows) in the training set.                                               |
| **Single training example**     | ( (x, y) )             | A pair containing one input and its corresponding output.                                             |
| **i-th training example**       | ( (x^{(i)}, y^{(i)}) ) | The *i-th* data point in the dataset (for example, the *i-th* row in your table).                     |

## Understanding Supervised Learning

In Supervised Learning, we train the model using data that already includes the correct answers (labels).
That means we provide the algorithm with both:

Features (x): Inputs

Targets (y): Known outputs

The model then learns the relationship between x and y.

## The Learning Process

Training Set → Learning Algorithm → Model (Hypothesis)

The learning algorithm takes the training set (features + targets) and produces a function (also called a hypothesis or model).

Model → Prediction

Once the model is trained, we can input a new feature value (x).

The model will output a prediction, written as y^ (read as y-hat):

y^ = f(x)

Here, y^ is the model’s estimate or prediction of y — not the actual value, but the model’s best guess.


## For Linear Regression

In Linear Regression, the hypothesis (model) is represented as:
y^ = fw,b(x)=w⋅x+b

Where:

w: Weight (or slope) — controls how much y changes with x

b: Bias (or intercept) — shifts the line up or down

x: Input feature

y^: Predicted output (y-hat)

#### Note:
You can think of w and b as the parameters the algorithm learns during training.
They determine how the line fits the data.
If w changes, the slope of the line changes.
If b changes, the position of the line shifts up or down.

# Problem Statement
In this lab we will use dataset with only two points - a house with 1000 square feet sold for 300,000 dollars and a house with 2000 square feet sold for 500,000 dollars. This two points will constitute our data or training set so the units of size are 1000sqft and the units of price are 1000s of dollars.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# x_train is the input variable (size in 1000 square feet)
# y_train is the target (price in 1000s of dollars)
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
print(f"x_train value is {x_train}")
print(f"y_train value is {y_train}")
```

```python
# Let's take the number of training examples - m
m = len(x_train)
print(f"Number of training examples are: {m}")
```

## Training Examples: x_i, y_i

We will use (x^(i),y^(i)) to denote the ith training example. 

Since python is zero indexed we will start with: 

(x^(0),y^(0)) is (2.0, 300.0) and 

(x^(1),y^(1)) is (3.0, 500.0)

```python
i = 0 

x_i = x_train[i]
y_i = y_train[i]
print(f"(x^({i}), y^({i})) = ({x_i}, {y_i})")

i = 1 

x_i = x_train[i]
y_i = y_train[i]
print(f"(x^({i}), y^({i})) = ({x_i}, {y_i})")
```

```python
# Plotting the data points
plt.scatter(x_train, y_train, marker = "x", c = "r")
plt.title("House price Prediction")
plt.xlabel('Size of houses (1000 sq ft)')
plt.ylabel('Price of houses (1000s dollars)')
plt.show()
```

# Model Function 
The model function for linear regression is :

y^ = fw,b(x^(i)) = w⋅x^(i)+b

The formula above is how you can represent straight lines - different values of w and b give you different staright lines on the plot.

```python
w = 200
b = 100
```

```python
def compute_model_output(x, w, b):
    m = x_train.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
    return f_wb
```

```python
output = compute_model_output(x_train, w, b)
plt.plot(x_train, output, c='b', label = "Our Prediction")
plt.scatter(x_train, y_train, marker = "x", c = "r")
plt.title("House Price Prediction")
plt.xlabel("Size(1000 sq ft)")
plt.ylabel("Price of houses(1000s dollars)")
plt.show()
```
